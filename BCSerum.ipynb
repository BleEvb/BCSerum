```python
#import modules

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#create dataframe and read excel file with pandas library

df_serums = pd.read_excel('BCSerum.xlsx', sheet_name='CSerum')
```

```python
#display dataframe before creating chart

df_serums
```

```python
#create and display plot chart for price

plt.figure(figsize = (9,6))
plt.bar(x = df_serums["Serum"],
        height = df_serums["Price £"],
        color = ['r','g','b','w','y'])
plt.xticks(rotation = 90, fontsize = 13)
plt.yticks(fontsize = 13)
plt.title("Battle of the C Serums", fontsize = 18, fontweight = "bold")
plt.ylabel("Price £", fontsize = 13)
plt.show()
```

```python
#create and display plot chart for rating

plt.figure(figsize = (9,6))
plt.bar(x = df_serums["Serum"],
        height = df_serums["Rate"],
        color =['r','g','b','w','y'] )
plt.xticks(rotation = 90, fontsize = 13)
plt.yticks(fontsize = 13)
plt.title("Battle of the C Serums", fontsize = 18, fontweight = "bold")
plt.ylabel("Rate/Stars", fontsize = 13)
plt.show()
```
